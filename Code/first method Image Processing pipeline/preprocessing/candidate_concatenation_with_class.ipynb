```python
def concatenate_csv_files(directory, output_file=None):
    # List to hold DataFrames
    dataframes = []

    csv_files = os.listdir(directory)
    csv_files = [csv for csv in csv_files if csv.endswith(".csv")]
    print(csv_files)
    # Iterate over all files in the directory
    for i, filename in enumerate(csv_files):
        print(f"[READ-IN] - {filename} - {i+1} out of {len(csv_files)}")

        # Construct full file path
        filepath = os.path.join(directory, filename)

        # Read the CSV file into a DataFrame
        df = pd.read_csv(filepath)

        # Extract the new value for the first column from the file name
        new_first_column_value = filename.replace('.csv', '')

        df = df.rename(columns={'Unnamed: 0': 'seriesuid'})

        # Replace the first column values with the new value
        df.iloc[:, 0] = new_first_column_value

        # Append the DataFrame to the list
        dataframes.append(df)

    # Concatenate all DataFrames into one
    concatenated_df = pd.concat(dataframes, ignore_index=True)

    if output_file is not None:
      # Save the concatenated DataFrame to a new CSV file
      concatenated_df.to_csv(output_file, index=False)

    return concatenated_df

# Example usage:
subset_candidates = concatenate_csv_files(SUBSET_PATH)
```
